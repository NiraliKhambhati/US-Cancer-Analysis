```python
# Histogram for specific column

import matplotlib.pyplot as plt
Cancer_data['Total.Rate'].hist(bins= 20)
plt.title('Distribution of Total Cancer Rates')
plt.xlabel('Total Rate')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter Plots

plt.scatter(Cancer_data['Total.Rate'], Cancer_data['Total.Population'])
plt.title('Scatter Plot of Total Rate vs Total Population')
plt.xlabel('Total Rate')
plt.ylabel('Total Population')
plt.show()
```

```python
# Most of the states have populations below 50 million, with only a few exceptions having populations significantly higher. This could correspond to a few highly populous states.
# Outliers: The plot shows a couple of states with exceptionally high populations. These are interesting points for further investigation to see if they are outliers or data errors.
```

```python
import seaborn as sns
```

```python
#Setting Aesthetics for the plots

sns.set(style= "whitegrid")
```

```python
# Distribution of Total Cancer Rates across States

plt.figure(figsize = (12,8))
sns.barplot(x = 'Total.Rate', y = 'State', data = Cancer_data.sort_values('Total.Rate', ascending = False))
plt.title('Distribution of Total Cancer Rates Across States')
plt.xlabel('Total Cancer Rate')
plt.ylabel('State')
plt.show()

#Interpretation: 
#States like West Virginia, Maine, and Pennsylvania might require more in-depth study to determine the causes of high cancer rates.Factors could include industrial pollution, high smoking rates, genetic predispositions, or less access to healthcare which could contribute to these statistics.
#
```
